# Which company manufactured the most number of vehicles ?

```python
# Head of the dataset.
car.head(5)
```

```python
# Count of the Make column.
car['Make'].value_counts()
```

```python
# Loading the required liabries.
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bar-plot of top 5 company who manufactured most number of vehicle.
plt.figure(figsize=(10,5))
sns.barplot(x=list(car['Make'].value_counts()[0:5].keys()),y=list(car['Make'].value_counts()[0:5]),ec='black',lw=1)
plt.xlabel("Name of the Company",fontsize=10)
plt.ylabel("Number of Vehicle",fontsize=10)
plt.grid(True)
plt.show()
```

From the above we can say that, most of the vehicles are manufactured by Toyota.

# Which is the most longest vehicle ?

```python
# Top 5 records of the dataset.
car.head()
```

```python
# Sort the car dataset in descending order of the Lenght column.
car.sort_values(by=['Length'], ascending=False)
```

```python
# Create a datatset a.
a=car.sort_values(by=['Length'], ascending=False).head()
a
```

```python
# Extract only Model and Length column from the datatset.
a=a[['Model','Length']]
a
```

```python
# List of the names of vehicles.
list(a['Model'])
```

```python
# List of the length of the vehicle. 
list(a['Length'])
```

```python
# Bar-plot of top 5 longest vehicle.
plt.figure(figsize=(13,5))
color = ['black','red','green','orange','blue','limegreen','darkgreen','royalblue','navy']
sns.barplot(x=list(a['Model']),y=list(a['Length']),ec='black',lw=1, palette=color)
plt.xlabel("Name of the Vehicle",fontsize=12)
plt.ylabel("Length of Vehicle",fontsize=12)
plt.grid(True)
plt.show()
```

From the above we can say that, Silverado SSn is the longest vehicle.
